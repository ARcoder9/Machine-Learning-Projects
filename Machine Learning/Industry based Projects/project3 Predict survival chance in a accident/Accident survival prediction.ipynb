```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Accident_train.csv")
df
```

```python
for i in range(17) :
    print("The number of null values in column",i,"are :")
    nullCheck=df.iloc[:,i].isnull().sum()
    print(nullCheck)
```

```python
df=df.dropna() #used to remove the rows which contains the null values
df
```

```python
for i in range(17) :
    print("The number of null values in column",i,"are :")
    nullCheck=df.iloc[:,i].isnull().sum()
    print(nullCheck)
```

```python
 df.Hour_of_Collision.unique()
```

```python
df.Junction_Control.unique() # it is used to find out the unique values in the column
```

```python
df=df.reset_index(drop=True) #use to reset the index or make things in order
df
```

```python
dummies1=pd.get_dummies(df['Weekday_of_Collision'])  
dummies1 #it is used to remove the string values with the dummies values
         # machine learning model is not much familiar with the string values
```

```python
dummies2=pd.get_dummies(df['Policing_Area'])
dummies2
```

```python
df.drop(['Policing_Area','Weekday_of_Collision'],axis='columns',inplace=True)
df
```

```python
df=pd.concat([df,dummies1,dummies2],axis='columns')
df
```

```python
y=df['Collision_Severity']
y
```

```python
y.unique()
```

```python
x=df.drop(['Collision_Severity','Collision_Ref_No'],axis='columns')
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_predicted=dt.predict(x_test)
y_predicted
```

```python
print("The first five predicted values are:")
print(y_predicted[0:5])
print("The first five test values are: ")
print(np.array(y_test[0:5]))
print(dt.score(x_test,y_test))
```

```python
from sklearn import metrics
matrix=metrics.confusion_matrix(y_predicted,y_test)
matrix
```

```python
sns.heatmap(matrix,cmap='Oranges',annot=True, fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Values")
plt.title("Confusion Matrix for DecisionTreeClassifier")
plt.show()
```
